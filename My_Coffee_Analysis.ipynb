```sql
-- Dimension table: customers  Q1
CREATE TABLE IF NOT EXISTS MY_COFFEE_SHOP_V1.dim_customers
(
    customer_id     BIGINT PRIMARY KEY   NOT NULL,
    first_name      VARCHAR(500)         NOT NULL,
    last_name       STRING               NOT NULL,
    email           STRING,
    tier            STRING,
    points          INT,                  
    created_at      DATE
);

--Create the dim_products table  Q2
CREATE TABLE IF NOT EXISTS MY_COFFEE_SHOP_V1.dim_products
(
product_id BIGINT PRIMARY KEY NOT NULL,
 product_code STRING NOT NULL,
 product_name STRING NOT NULL,
 price DECIMAL(10,2) NOT NULL,
category STRING,
 is_active BOOLEAN  
);



--Create the fact_orders table, partitioned by order_date  Q3
CREATE TABLE IF NOT EXISTS MY_COFFEE_SHOP_V2.fact_orders
(
    order_id      BIGINT PRIMARY KEY,
    customer_id   BIGINT NOT NULL,
    product_id    BIGINT NOT NULL, 
    store_id      BIGINT NOT NULL,
    order_date    DATE,
    quantity      INT,
     unit_price   DECIMAL(10,2),
     total_amount DECIMAL(10,2)
)
PARTITIONED BY (order_date);

--Enable column mapping on dim_customers
ALTER TABLE MY_COFFEE_SHOP_V1.dim_customers
SET TBLPROPERTIES (
    'delta.columnMapping.mode' = 'name',
    'delta.minReaderVersion' = '2',
    'delta.minWriterVersion' = '5'
);
```

Back up dim_customers, then delete one customer by their customer_id.
REQUIRED STEPS: CREATE TABLE … AS SELECT SELECT (verify row) DELETE … WHERE customer_id = ? DESCRIBE HISTORY
★ Always back up before deleting. Delete by customer_id only - never by name. Check DESCRIBE HISTORY to see the delete in the Delta log.

```sql
-- Step 1: Placing Bob back into the main table with his original data
INSERT INTO MY_COFFEE_SHOP_V1.dim_customers (customer_id, first_name, last_name, email, tier, points, created_at)
VALUES (4, 'Bob', 'Williams', 'new_bob_email@gmail.com', 'gold', 1500, '2022-01-04');

-- Step 2: overwrite the backup table so he is safely stored there
CREATE OR REPLACE TABLE MY_COFFEE_SHOP_V1.dim_customers_backup AS 
SELECT * FROM MY_COFFEE_SHOP_V1.dim_customers;

-- Step 3: delete by ID
DELETE FROM MY_COFFEE_SHOP_V1.dim_customers WHERE customer_id = 4;

-- Step 4: confirm
SELECT * FROM MY_COFFEE_SHOP_V1.dim_customers;

```
